```python
# Reading unicode data
unicode_data = []
with open ('/content/drive/MyDrive/Colab Notebooks/NLP/unicode','r') as File:
  for i in File:
    i = re.sub(r' ', r'', i)
    i = re.sub(r'([^\s])', r'\1 ', i)
    #print(i)
    unicode_data.append(i)
File.close()
```

```python
# Reading zawgyi data
zawgyi_data = []
with open ('/content/drive/MyDrive/Colab Notebooks/NLP/zawgyi.zawgyi','r') as File:
  for i in File:
    i = re.sub(r' ', r'', i)
    i = re.sub(r'([^\s])', r'\1 ', i)
    #print(i)
    zawgyi_data.append(i)
File.close()
```

```python
# Train data of unicode_data 70%  & Test data of unicode_data 30%
len(unicode_data) * 0.7
train_unicode_data = unicode_data[0:7700]
test_unicode_data = unicode_data[7700:]

# Train data of zawgyi_data 70%  & Test data of zawgyi_data 30%
len(zawgyi_data) * 0.7
train_zawgyi_data = zawgyi_data[0:7700]
test_zawgyi_data = zawgyi_data[7700:]
```

```python
# Labeling unicode : 1 & zawgyi : 0
train = train_unicode_data + train_zawgyi_data
train_labels = [1]*len(train_unicode_data) + [0]*len(train_zawgyi_data)
test = test_unicode_data + test_zawgyi_data
test_labels = [1]*len(test_unicode_data) + [0]*len(test_zawgyi_data)
```

```python
# labeling randomly (frequency ne' count myar tar ayin label )
oov_tok = '<OOV>' # out of vocab
sample = ['I am Nu Nu Hlaing']
tokenizer = Tokenizer(oov_token = oov_tok)
tokenizer.fit_on_texts(sample) # data char 1ku chin si number tat
word_index = tokenizer.word_index
word_index
```

```python
# change to sequences
sequences = tokenizer.texts_to_sequences(sample)
sequences
```

```python
max_length = 4182
trunc_type = 'post'  #put needed '0's for max length , if exceed, delete
vocab_size = 500  # firstmost 500 char ko be' frequency count
padded = pad_sequences(sequences,maxlen=max_length, truncating=trunc_type,padding='post') # length standard thet mhat
padded
```

```python
#
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(train)
train_padded = pad_sequences(sequences,maxlen=max_length, truncating=trunc_type,padding='post')
train_padded
```

```python
embedding_dim = 64
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(1000, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid') #output layer sigmoid fun: ko use htar loz 0～1 output htwt mhr
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
type(train_labels)
```

```python
num_epochs = 10
history = model.fit(train_padded, np.array(train_labels), epochs=num_epochs,verbose = 2)
```

```python
model.save('/Unicode_Model.h5')
```

# 5. Evaluation

```python
testing_sequences = tokenizer.texts_to_sequences(test)
testing_padded = pad_sequences(testing_sequences,maxlen=max_length, truncating=trunc_type,padding='post')
testing_padded
```

```python
predicted_results = model.predict(testing_padded)
predicted_results
```

```python
result = []
for i in predicted_results:
    if(i<0.5):
        result.append(0)
    else:
        result.append(1)
```

```python
#result
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(test_labels, result)
print('confusion_matrix')
print(confusion_matrix)
```

```python
# Evaluation Metrics
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score

print('Accuracy Score',accuracy_score(test_labels, result)*100,'%')
print('Precision Macro Score',precision_score(test_labels, result, average = 'macro')*100,'%')
print('Recall_Score',recall_score(test_labels, result, average = 'macro')*100,'%')
print('F1_Score',f1_score(test_labels, result, average = 'macro')*100,'%')
```
